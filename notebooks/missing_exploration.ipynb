# Intro

This notebook explores the missingness structure of Understanding Society data for wave 21 (2011-2012).

To begin load data and count number of missing values by column. There are 5 missing types

-1 Don't Know - An individual does not know the answer to the question (e.g. birth year)
-2 Refusal - An individual refuses to answer a question (e.g. political opinion)
-7 Proxy - An individual was unable to attend their interview. A shorter proxy interview was undertaken by mail/next of kin with some questions not answered.
-8 Inapplicable - An observation does not have a logical answer by design. E.g. Anyone unemployed will not have an income variable.  For US NA also includes observations not carried forwards. Some variables are not recorded unless an individual changes state. They aren't technically missing. distinction between 'truly' NA values and tbose not carried forwards may be done by assigning missing by design values -10.
-9 Missing (unintentionally) - an observation is truly missing. This can be for any number of data recording errors such as human error.
